```python
import pandas as pd
import numpy as np
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
import sklearn
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
df = pd.read_csv("IMDB Dataset.csv")
df
```

```python
df = df.drop_duplicates()
```

```python
df.isnull().any()
```

```python
df = df.apply(lambda x: x.astype(str).str.lower() if x.dtype == "object" else x)
```

```python
df["review"] = df["review"].str.replace('[^a-zA-Z0-9 ]','')
```

```python
df["review"] = df["review"].str.replace('br','')
```

```python
df['sentiment'] = df['sentiment'].replace({'positive': 1, 'negative': 0})
df
```

# Model

```python
X = list(df["review"])
```

```python
y = list(df["sentiment"])
```

```python
from transformers import BertTokenizer, TFBertModel, BertConfig, TFBertForSequenceClassification, DistilBertTokenizerFast

tokenizer = DistilBertTokenizerFast.from_pretrained('distilbert-base-uncased')
```

```python
#Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 48)
```

```python
#Encoding
train_encoding = tokenizer(X_train, truncation=True, padding=True)
test_encoding = tokenizer(X_test, truncation=True, padding=True)
```

```python
#Turning encodings into dataset
train_df = tf.data.Dataset.from_tensor_slices((
    dict(train_encoding),
    y_train
))

test_df = tf.data.Dataset.from_tensor_slices((
    dict(test_encoding),
    y_test
))
```
